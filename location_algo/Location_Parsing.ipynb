```python
#After loading the link: gs://mic-check-ai/scripts/tweet_location_files/test.json
with open('test.json') as data_file:    
    data = json.load(data_file)  
```

```python
df = pd.DataFrame(data).transpose()
```

```python
location_dic = dict()
```

```python
#This loops through each region/city & creates a dictionary that has the city with the hours in which the tweets were created
for i in range(df.shape[0]):
     location_dic[df.iloc[i].name]= np.array(pd.to_datetime(pd.DataFrame(df.iloc[i]['statuses']).transpose()['created_at']).dt.hour)
```

```python
#Creates a data frame from the above dictionary
df_hours = pd.DataFrame.from_dict(location_dic,orient='index').transpose()
```
